# Rectified Linear Unit- Relu 

f(x) = max(0, x)


```python
import numpy as np
import matplotlib.pyplot as plt

def ReLU(x):
    return np.maximum(0, x)

def dReLU(x):
    dev = np.zeros(x.shape[0])
    for i in range(x.shape[0]):
        if x[i] > 0:
            dev[i] = 1
    return dev

# ReLU
plt.figure()

x = np.linspace(-10, 10, 100)

plt.plot(x, ReLU(x), 'b', label='linspace(-10, 10, 10)')

plt.grid()

plt.title('ReLU Function')

plt.text(2.5, 0.5, r'$f(x)=max(0, x)$', fontsize=15)

plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))

plt.xlim((-10, 10))

# dReLU
plt.figure()
x = np.linspace(-10, 10, 100)

plt.plot(x, dReLU(x), 'b', label='linspace(-10, 10, 10)')

plt.grid()

plt.title('dReLU Function')

plt.text(2.5, 0.5, r'$f\'(x)=u(x)$', fontsize=15)

plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))

plt.xlim((-10, 10))

plt.show()
```
